```python
# Import pandas and matplotlib libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Name of the method 
#Method = "Lax"
#Method = "Lax Wendroff"
#Method = "MacCormack"
Method = "RungeKutta"

name1 = "Error for the " + Method + " method.csv"
name2 = "Numerical result for " + Method + " method.csv"
name3 = "Analytical result for " + Method + " method.csv"

# Read the csv file into a dataframe
df_error = pd.read_csv(name1)
df_numerical = pd.read_csv(name2)
df_analytical = pd.read_csv(name3)
```

```python
#Plotting error of the simulation 
plt.plot(df_error["time"], df_error["error"])
min(df_error["error"])
plt.xlabel("Iterations")
plt.ylabel("Error in percent")
plt.title("Error of analytic and numeric result")

```

```python
x = df_analytical["Position_x"]
y = df_analytical["Position_y"]
u_analytic = df_analytical["value"]

#changing data to matrix
Nx = int(np.sqrt(x.size))
Ny = int(np.sqrt(y.size))

x = np.array(x)
x.resize(Ny, Nx)

y = np.array(y)
y.resize(Ny, Nx)

u_analytic = np.array(u_analytic)
u_analytic.resize(Ny, Nx)

plt.contourf(x, y, u_analytic, 100)
plt.colorbar()
```

```python
x = df_numerical["Position_x"]
y = df_numerical["Position_y"]
u_numeric = df_numerical["value"]

#changing data to matrix
Nx = int(np.sqrt(x.size))
Ny = int(np.sqrt(y.size))

x = np.array(x)
x.resize(Ny, Nx)

y = np.array(y)
y.resize(Ny, Nx)

u_numeric = np.array(u_numeric)
u_numeric.resize(Ny, Nx)

plt.contourf(x, y, u_numeric)
plt.colorbar()
```

```python
plt.contourf(x, y, u_numeric, vmin = -0.5, vmax = 1.5)
plt.clim(-0.5, 1.5)
plt.colorbar()
plt.show()
```
